## Ajustes de columnas

```python
#Eliminamos gross_revenue porque no está bien calculada
df = df.drop(columns=["gross_revenue"])
```

```python
df = df.drop(columns=["profit"])
```

```python
#Volvemos a crearla bien calculada
df["gross_revenue"] = df["unit_price"] * df["quantity"] * (1 - df["discount_pct"])
```

```python
df["profit"] = (
    df["gross_revenue"]
    - df["unit_cost"] * df["quantity"]
    - df["shipping_cost"]
)
```

# VENTAS Y TENDENCIA

```python
total_revenue = df["gross_revenue"].sum()
total_profit = df["profit"].sum()

print(total_revenue, total_profit)
```

```python
profit_margin = total_profit / total_revenue
profit_margin
```

```python
total_orders = df["order_id"].nunique()

ticket = df.groupby("order_id")["gross_revenue"].sum()
avg_ticket = ticket.mean()

total_orders, avg_ticket
```

```python
df["year_month"] = df["order_date"].dt.to_period("M")
```

```python
monthly_sales = df.groupby("year_month")["gross_revenue"].sum()

monthly_sales
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

monthly_sales.plot(marker='o')

plt.title("Evolución de Ingresos Mensuales", fontsize=14)
plt.xlabel("Mes")
plt.ylabel("Revenue (€)")

plt.grid(alpha=0.3)

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
monthly_orders = df.groupby("year_month")["order_id"].nunique()
monthly_orders
```

```python
monthly_ticket = df.groupby("year_month")["gross_revenue"].sum() / monthly_orders
monthly_ticket
```

# PRODUCTOS Y CATEGORIAS

Top 10 productos más vendidos

```python
top_productos_volumen = (
    df.groupby("product")["quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_productos_volumen
```

Productos que generan más ingresos

```python
top_productos_revenue = (
    df.groupby("product")["gross_revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_productos_revenue
```

Productos que generan mas devoluciones

```python
tasa_devolucion_producto = (
    df.groupby("product")["returned_flag"].mean().sort_values(ascending=False)
)

print(tasa_devolucion_producto.head(10)*100)
```

Distribución de las ventas entre las diferentes categorías
